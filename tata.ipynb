```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_excel(r"C:\Users\user\Downloads\Coursera MASKED data.xlsx")
```

```python
# Remove rows where the 'Completed' column is blank
cleaned_df = df.dropna(subset=['Completed'])

# Convert 'Completed' to a consistent format (e.g., all lowercase)
cleaned_df['Completed'] = cleaned_df['Completed'].str.lower()

# Calculate the total count and the 'yes' count for each division
division_total = cleaned_df['Division'].value_counts()
division_yes = cleaned_df[cleaned_df['Completed'] == 'yes']['Division'].value_counts()

# Calculate the percentage of 'yes' completions for each division
division_percentage = (division_yes / division_total) * 100

# Get the top 10 divisions by percentage of 'yes' completions
top_10_divisions = division_percentage.nlargest(23)

# Plot the data using horizontal bars
fig, ax = plt.subplots(figsize=(12, 6))
bars = top_10_divisions.plot(kind='barh', color=plt.cm.tab20.colors[1:11], ax=ax)

# Set the full background color
fig.patch.set_facecolor('#0071C9')
ax.set_facecolor('#0071C9')

# Add labels to each bar
for bar in bars.patches:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f}%', va='center', ha='left', color='white')

# Set labels and title to white
ax.set_xlabel('Percentage Completed', color='white')
ax.set_ylabel('Division', color='white')
ax.set_title('Divisions by Percentage of Completion', color='white')

# Set tick params to white
ax.tick_params(colors='white')

# Invert y-axis to show the highest percentage at the top
plt.gca().invert_yaxis()

# Tight layout
plt.tight_layout()

output_path = 'top_10divisions.png'
plt.savefig(output_path, bbox_inches='tight')

# Display the plot
plt.show()

print(f"Graph saved as {output_path}")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.table import Table

df= pd.read_excel(r"C:\Users\user\Downloads\Coursera MASKED data.xlsx")

# Remove rows where the 'Completed' column is blank
cleaned_df = df.dropna(subset=['Completed'])

# Convert 'Completed' to a consistent format (e.g., all lowercase)
cleaned_df['Completed'] = cleaned_df['Completed'].str.lower()

# Calculate total number of courses per division
total_courses = cleaned_df.groupby('Division').size()

# Calculate number of completed courses (yes) per division
completed_courses = cleaned_df[cleaned_df['Completed'] == 'yes'].groupby('Division').size()

# Calculate the average course grade per division
average_course_grade = cleaned_df.groupby('Division')['Course Grade'].mean()

# Create a summary dataframe
summary_df = pd.DataFrame({
    'Total Courses Opted': total_courses,
    'Completed Courses': completed_courses,
    'Average Course Grade': average_course_grade
})

# Fill NaN values with 0
summary_df = summary_df.fillna(0)

# Create the plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_axis_off()
fig.patch.set_facecolor('#0071C9')

# Create the table
tbl = Table(ax, bbox=[0, 0, 1, 1])
tbl.auto_set_font_size(False)
tbl.set_fontsize(12)

# Add cells to the table
n_rows, n_cols = summary_df.shape
header_color = '#0071C9'
row_colors = ['#0071C9'] * (n_rows + 1)
edge_color = 'white'
text_color = 'white'

# Add header row
for j in range(n_cols + 1):
    if j == 0:
        tbl.add_cell(0, j, width=1/n_cols, height=1/(n_rows+1), text='Division', loc='center', facecolor=header_color, edgecolor=edge_color)
    else:
        tbl.add_cell(0, j, width=1/n_cols, height=1/(n_rows+1), text=summary_df.columns[j-1], loc='center', facecolor=header_color, edgecolor=edge_color)

# Add data rows
for i in range(n_rows):
    for j in range(n_cols + 1):
        if j == 0:
            value = summary_df.index[i]
        else:
            value = summary_df.iloc[i, j-1]
        tbl.add_cell(i+1, j, width=1/n_cols, height=1/(n_rows+1), text=value, loc='center', facecolor=row_colors[i], edgecolor=edge_color)

# Update text properties for the entire table
for key, cell in tbl.get_celld().items():
    cell.set_text_props(color=text_color)

# Add the table to the plot
ax.add_table(tbl)

# Save the table as an image
plt.tight_layout()
image_path = 'summary_table.png'
plt.savefig(image_path, dpi=300, bbox_inches='tight')
plt.show()

# Output the path of the saved image
print(f"Image saved at: {image_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
df= pd.read_excel(r"C:\Users\user\Downloads\Coursera MASKED data.xlsx")


# Ensure 'Learning Hours' column is numeric

df['Duration(in hrs)'] = pd.to_numeric(df['Duration(in hrs)'], errors='coerce')

# Remove rows with NaN values in 'Learning Hours'
df = df.dropna(subset=['Duration(in hrs)'])
distinct_courses_df = df.loc[df.groupby('Course Name')['Duration(in hrs)'].idxmax()]
# Get the top 10 courses based on 'Learning Hours'
top_10_courses_by_duration = distinct_courses_df.nlargest(10, 'Duration(in hrs)')[['Course Name', 'Duration(in hrs)']]

# Plotting the top 10 courses based on duration
fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#0071C9')
ax.set_facecolor('#0071C9')

# Generate a list of different colors for the bars
colors = plt.cm.tab20.colors[2:11]

# Plot the horizontal bar chart
bars = ax.barh(top_10_courses_by_duration['Course Name'], top_10_courses_by_duration['Duration(in hrs)'], color=colors)

# Add labels to each bar
for bar in bars:
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height()/2, f'{width:.2f}', va='center', ha='left', color='white')

# Set the title and labels
ax.set_title('Top 10 Largest Duration Courses', color='white', fontsize=14)
ax.set_xlabel('Duration(in hrs)', color='white')
ax.set_ylabel('Course Name', color='white')

# Customize tick parameters
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Tight layout for better spacing
plt.tight_layout()

# Save and display the plot
image_path = 'top_10_courses_by_duration.png'
plt.savefig(image_path, dpi=300, bbox_inches='tight')
plt.show()

# Output the path of the saved image
print(f"Image saved at: {image_path}")
```
